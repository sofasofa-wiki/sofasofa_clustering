```python
import os 
import json 

import numpy as np
```

```python
path = '../data/sofasofa-data'

descs = []
for _, _, c in os.walk(path):
    for name in c:
        if name != '.DS_Store':
            file_path = os.path.join(path, name)
            with open(file_path, 'r', encoding='utf-8') as fr:
                jdata = json.load(fr)
                descs.append(jdata['description'])
```
